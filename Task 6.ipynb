### Importing necessary libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### Loading the Iris dataset

```python
df = pd.read_csv('Iris.csv')
df.head()
```

### Encoding the labels
-Converting Categorical data into numbers for predictive models to perform task.

```python
le_species = LabelEncoder()
df['Species_n'] = le_species.fit_transform(df['Species'])
df.head()
```

```python
inputs = df.drop(['Species', 'Species_n', 'Id'], axis=1)
target = df['Species_n']
```

```python
inputs.head()
```

```python
target.head()
```

### Splitting the data into training and testing sets by using Scikit-Learn's built-in train_test_split() method

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=0)
```

### Defining the decision tree algorithm 

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model = model.fit(X_train, y_train)
```

```python
print('Decision tree classifier created')
```

```python
model.score(X_train, y_train)
```

### Prediction Making

```python
model.predict([[6.8, 0.5, 3.4, 5.2]])
```

### Visualizing the decision tree

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(30,20))
tree.plot_tree(model, filled=True, feature_names=inputs.columns, rounded=True)
plt.show()
```
